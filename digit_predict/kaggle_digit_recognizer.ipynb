```python
def binarysation(dataset):
    col_id=1; # skip label col
    while col_id < dataset.shape[1]:
        for i,row in dataset.iterrows():
            
            if row[col_id] and col_id!=0 > 128:
                dataset.ix[i,col_id]=1;
            else:
                dataset.ix[i,col_id]=0;
        col_id+=1;
    return dataset
    
```
