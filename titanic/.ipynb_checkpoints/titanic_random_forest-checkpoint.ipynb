```python
import os
import pandas as pd
import sklearn
import numpy as np

```

```python
data_filename = os.path.join("data", "train.csv")
dataset = pd.read_csv(data_filename)
def clear_data_input(dataset):
    dataset.Age=dataset.Age.fillna(dataset.Age.mean()*2.5)
    dataset.SibSp=dataset.SibSp.fillna(0)
    dataset.Parch=dataset.Parch.fillna(0)
    dataset.Fare=dataset.Fare.fillna(8)
    dataset.Embarked=dataset.Embarked.fillna("S")
    for i,row in dataset.iterrows():
        if row["Embarked"]=="C":
            dataset.ix[i,"Embarked"]=1;
        elif  row["Embarked"]=="S":
            dataset.ix[i,"Embarked"]=2;
        elif  row["Embarked"]=="Q":
            dataset.ix[i,"Embarked"]=0; 
        if row.Sex=="male":
            dataset.ix[i,"Sex"]=1;
        else: 
            dataset.ix[i,"Sex"]=0; 
    return dataset
def add_name_index(dataset):
    mapping={' Mr':1,
 ' Mrs': 2,
 ' Miss': 3,
 ' Master': 4,
 ' Don':1,
 ' Rev':6,
 ' Dr':5,
 ' Mme':8,
 ' Ms': 2,
 ' Major': 7,
 ' Lady': 10,
 ' Sir': 9,
 ' Mlle': 8,
 ' Col': 7,
 ' Capt': 7,
 ' the Countess': 10,
 ' Jonkheer': 10,
 ' Dona': 2}
    for i,row in dataset.iterrows():
        abr=row["Name"].split(',')[1].split('.')[0];
        dataset.ix[i,"Name_Index"]=mapping[abr];
                
    return dataset   
```

```python
dataset=clear_data_input(dataset)
dataset=add_name_index(dataset)

dataset.ix[:10]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
features=["Pclass","Sex","Age","Embarked","Fare","SibSp","Parch","Name_Index"]
clf = RandomForestClassifier(random_state=14)
scores = cross_val_score(clf, dataset[features], dataset["Survived"], scoring='accuracy')
print("Using full team labels is ranked higher")
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
from sklearn.grid_search import GridSearchCV
parameter_space = {
                   "max_features": [3,5,7, 'auto'],
                   "n_estimators": [100,],
                   "criterion": ["gini", "entropy"],
                   "min_samples_leaf": [2, 4, 6],
                   }
clf = RandomForestClassifier(random_state=14)
grid = GridSearchCV(clf, parameter_space)
grid.fit(dataset[features], dataset["Survived"])
print("Accuracy: {0:.1f}%".format(grid.best_score_ * 100))
print(grid.best_estimator_)
```

```python
samples_filename = os.path.join("data", "test.csv")
samples = pd.read_csv(samples_filename)
samples = clear_data_input(samples)
samples = add_name_index (samples)

features=["Pclass","Sex","Age","Embarked","Fare","SibSp","Parch","Name_Index"]
clf_final=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='entropy',
            max_depth=None, max_features=5, max_leaf_nodes=None,
            min_impurity_split=1e-07, min_samples_leaf=4,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            n_estimators=100, n_jobs=1, oob_score=False, random_state=14,
            verbose=0, warm_start=False)
clf_final.fit(dataset[features], dataset["Survived"])
predictions=clf_final.predict_proba(samples[features])[:,1]
predictions[predictions<= .5] =0;
predictions[predictions > .5] =1;
predictions.astype(int)
submission = pd.DataFrame ({
        "PassengerId": samples["PassengerId"],
        "Survived": predictions.astype(int)
    })
submission.to_csv("kaggle.csv", index=False)
print (submission);
```
